````python
# MY OWN EXPERIMENTS

# Example lists
brands = ["Toyota", "BMW", "Tesla", "Audi", "Toyota", "Tesla", "BMW"]
prices = [20000, 35000, 50000, 40000, 22000, 52000, 36000]


# Total number of cars sold
print(f"Cars sold = {len(brands)}\n")


# Unique brands
unique_brands = list(set(brands))
print(f"List of brands: {unique_brands}\n")


# Cars and price
print("====== CARS CATALOG ======")
for i, brand in enumerate(brands):  # Get both the item and the index
  print(f"Brand: {brand}, Price: ${prices[i]}")
print()


# Count cars by price
print("====== COUNT BY BRAND ======")
for brand in unique_brands:
  print(f"Brand {brand} = {brands.count(brand)} cars available")
print()


# Maximum and minimum price
print("====== MIN/MAX PRICES ======")
# Get the highest/lowest price
min_price = min(prices)
max_price = max(prices)

# Get the position of that prices
min_price_index = prices.index(min_price)
max_price_index = prices.index(max_price)

# Get the car brand on that position
cheapest_car = brands[min_price_index]
expensive_car = brands[max_price_index]

print(f"Cheapest car: {cheapest_car}")
print(f"Expensive car: {expensive_car}")


# List expensive cars (over 40000) using list comprehension


````

### Drawing charts
Library `Matplotlib`

````python
import matplotlib.pyplot as plt
````

Let us begin with a simple one, without frills.

````python
x = np.linspace(start=-2, stop=2, num=100)
y = x**2
plt.plot(x,y);
````

And now more detailed charts, suitable for reports, articles, diploma thesis, etc.

````python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10,4))

x = np.arange(start=0, stop=2, step=0.1)
y = x**2
ax[0].plot(x,y, marker='.')
ax[0].set_xlabel('x', fontsize=14)
ax[0].set_ylabel('y', fontsize=14)

x = np.linspace(start=0, stop=6, num=100)
y = np.sin(x) + np.sin(3*x)
ax[1].plot(x,np.sin(x), label='component 1', alpha=0.8, linestyle='dashed')
ax[1].plot(x,np.sin(3*x), label='component 2', alpha=0.8, linestyle='dotted')
ax[1].plot(x,y, label='sum', linewidth=3)
ax[1].set_xlabel('x', fontsize=14)
ax[1].set_ylabel('y', fontsize=14)
ax[1].legend()
fig.tight_layout()
plt.show()
````

Laboratory of waves

````python
t = np.linspace(0,2*np.pi, 600)
x = np.sin(3*t)
y = np.cos(5*t)

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11,4))

ax[0].plot(t,x, label='x')
ax[0].plot(t,y, label='y')
ax[0].set_xlabel('t', fontsize=14)
ax[0].set_ylabel('value', fontsize=14)
ax[0].legend(fontsize=12)
ax[0].set_title('time plot')

ax[1].plot(x,y)
ax[1].set_xlabel('x', fontsize=14)
ax[1].set_ylabel('y', fontsize=14)
ax[1].axis('square')
ax[1].set_title('state space plot')

fig.tight_layout()
plt.show()
````
